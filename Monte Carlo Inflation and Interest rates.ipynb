```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define the variables
inflation_data = [1.8,1.5,1.6,1.1,1.5,2.1,2.4,1.8,1.2,5.4,8.0,7.0] # Historical inflation data
interest_rates = [0.50, 0.75, 1.0,1.25,1.50,1.75,2.0,2.25,2.50,3.0,3.25,3.5,3.75,4.0,4.25]
simulations = 1000 # Number of simulations

# Create a function to simulate the rate increase
def rate_increase(inflation_data, interest_rate):
    inflation = np.random.choice(inflation_data)
    if interest_rate > inflation:
        return 1
    else:
        return 0

# Run the simulation
results = np.zeros((len(interest_rates), simulations))
for i in range(len(interest_rates)):
    for j in range(simulations):
        results[i][j] = rate_increase(inflation_data, interest_rates[i])

# Calculate the probability of rate increase for each interest rate
probabilities = np.mean(results, axis=1)

# Plot the results
plt.figure(figsize=(10,6))
plt.plot(interest_rates, probabilities)
plt.xlabel('Interest Rate')
plt.ylabel('Probability of Rate Increase')
plt.title('Monte Carlo Simulation for Rate Increase')
plt.show()
```
