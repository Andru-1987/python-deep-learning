```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import keras
from keras.utils.np_utils import to_categorical
from keras.layers import Dense
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.optimizers import SGD
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.layers import LeakyReLU
from keras.utils import plot_model

theta = np.linspace(0,2*np.pi,100) 
R=1-np.sin(theta)    
x = R*np.cos(theta)
y = R*np.sin(theta)

coordinates_df=pd.DataFrame({'x':x,'y':y})

print(coordinates_df)

scaler = StandardScaler()

scaler.fit(coordinates_df)

StandardScaler(copy=True, with_mean=True,with_std=True)
coordinates_df=scaler.transform(coordinates_df)


model = Sequential()

model.add(Dense(100, input_shape=(2,)))
model.add(LeakyReLU(alpha=0.03))
# Add two Dense layers with 50 neurons and relu activation
model.add(Dense(100))
model.add(LeakyReLU(alpha=0.03))

# End your model with a Dense layer and no activation
model.add(Dense(1))

# Compile your model
model.compile(optimizer='sgd', loss='mean_squared_error', metrics=['accuracy'])

plot_model(model, to_file='model.png')
img=plt.imread('model.png')
plt.imshow(img)
plt.show()


plt.figure(figsize=(10,3)) #fig size same as before
ax = plt.gca() #you first need to get the axis handle
ax.set_aspect(1.5) #sets the height to width ratio to 1.5. 
#Use help(ax.set_aspect) for more options.

ax.plot(x,y)
plt.show()

```

```python
# Fit your model on your data for 30 epochs
#X_train,y_train,X_test,y_test=train_test_split(coordinates_df,R,test_size=.3,random_state=42)

#print(R)
#print(y_train)

history=model.fit(coordinates_df,R, epochs = 600)
#history=model.fit(X_train,y_train,  epochs = 600)

# Evaluate your model 
print("Final lost value:",model.evaluate(coordinates_df,R))

# Plot the twenty minute orbit 
#plot_orbit(twenty_min_orbit)

```

```python
#print(pred_coordinates)
# Predict the twenty minutes orbit
pred_R = model.predict(coordinates_df)
#print(pred_R)
item_count=len(pred_R)
pred_x=[]
pred_y=[]
for i in np.arange(0,item_count):
    pred_x.append(pred_R[i]*np.cos(theta[i]))
    pred_y.append(pred_R[i]*np.sin(theta[i]))


plt.figure(figsize=(10,3)) #fig size same as before
ax = plt.gca() #you first need to get the axis handle
ax.set_aspect(1.5) #sets the height to width ratio to 1.5. 
ax.plot(pred_x,pred_y)
plt.show()
```
