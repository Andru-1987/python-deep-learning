```python
#https://aqs.epa.gov/aqsweb/airdata/download_files.html#Daily
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("nitrogen_dioxide_2020.csv")
```

```python
print(df.columns)
filter=df['State Name']=='Utah'
utah_aqi=df[filter]

print(utah_aqi.head())
```

```python
sns.relplot(x="1st Max Hour", y="1st Max Value",
    data=utah_aqi,
    kind="line")
plt.title("Nitrogen Oxiode levels in Utah")
plt.show()

sns.relplot(x=range(len(utah_aqi)), y="AQI",
    data=utah_aqi,
    kind="line")
plt.title("AQI levels in Utah")
plt.show()
```

```python
print(utah_aqi[utah_aqi['AQI']>100]["County Name"].head())
```

```python
# Air Quanity Index occurrences by county

sns.catplot(y="AQI", data=utah_aqi,
            kind="count", col="County Name")

# Show plot
plt.show()

print("Air Quality Index above 100 to 150 may cause individual with health issues to experience difficulties")
```
