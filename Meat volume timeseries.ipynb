```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#df=pd.read_csv('https://s3.amazonaws.com/assets.datacamp.com/production/course_5226/datasets/ch4_meat.csv')
#df.to_csv('ch4_meat.csv')

df=pd.read_csv('ch4_meat.csv',parse_dates=['date'],index_col='date')
print(df.columns)
plt.style.use('fivethirtyeight')
ax=df.plot.area(figsize=(12,4), fontsize=14)
plt.show()
```

```python

COLUMNS=['beef', 'veal', 'pork', 'lamb_and_mutton', 'broilers','other_chicken', 'turkey']
df_summary=df[COLUMNS].agg(['mean','sum'])

plt.clf()
ax=df[COLUMNS].plot(colormap='Dark2',figsize=(10,7))

ax.table(cellText=df_summary[COLUMNS].values,colWidths=[0.3]*len(df_summary[COLUMNS].columns),\
    rowLabels=df_summary.index,
    colLabels=df_summary[COLUMNS].columns,
    loc='top')

plt.show()
```

```python
COLUMNS=['beef', 'veal', 'pork', 'lamb_and_mutton', 'broilers','other_chicken', 'turkey']

ax=df[COLUMNS].plot(subplots=True,    linewidth=0.5, layout=(3,3),    figsize=(16,10),    sharex=False,    sharey=False)
#lt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
plt.show()
```

```python
corr_p = df[COLUMNS].corr(method='pearson')
#corr_p = df[COLUMNS].corr(method='spearman')
#print(corr_p)
cmap=sns.diverging_palette(h_neg=10, h_pos=240, as_cmap=True)
mask=np.triu(np.ones_like(corr_p,dtype=bool))
sns.heatmap(corr_p, center=0, cmap=cmap, linewidths=1,
annot=True, fmt=".2f",mask=mask)
plt.show()


sns.clustermap(corr_p,center=0, cmap=cmap, linewidths=1,
annot=True, fmt=".2f",mask=mask)
plt.show()
```
