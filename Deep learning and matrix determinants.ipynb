```python
import numpy as np
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers.merge import concatenate
from sklearn.model_selection import train_test_split
from keras.utils import plot_model
from matplotlib import pyplot as plt
import pandas as pd
from keras.callbacks import EarlyStopping
```

```python
#https://www.mathsisfun.com/algebra/systems-linear-equations-matrices.html

#https://www.mathsisfun.com/algebra/matrix-inverse.html

#https://www.wikihow.com/Find-the-Inverse-of-a-3x3-Matrix

# 4x + 3y + 2z = 25
# -2x + 2y + 3z = -10
# 3x -5y + 2z = -4

A = np.array([[4, 3, 2], [-2, 2, 3], [3, -5, 2]])
B = np.array([25, -10, -4])

inv_A = np.linalg.inv(A)

print(inv_A)
X=inv_A.dot(B)
print(X)
```

```python

input_tensor_1=Input(shape=(1,))
input_tensor_2=Input(shape=(1,))
input_tensor_3=Input(shape=(1,))
input_tensor_4=Input(shape=(1,))

merge=concatenate([input_tensor_1,input_tensor_2,input_tensor_3,input_tensor_4])

hidden1 = Dense(50, activation='relu')(merge)
hidden2 = Dense(50, activation='relu')(hidden1)
# prediction output
output = Dense(1)(hidden2)

model=Model([input_tensor_1,input_tensor_2,input_tensor_3,input_tensor_4],output)

model.compile(optimizer='sgd', loss='mean_absolute_error')

plot_model(model, to_file='model.png')
img=plt.imread('model.png')
plt.imshow(img)
plt.show()
```

```python

#df=pd.DataFrame({"A":A,"B":B,"X":X})
#print(A)
arr=A.flatten()
arr1=arr[0:3]
arr2=arr[3:6]
arr3=arr[6:9]
#print(arr1)
#print(arr2)
#print(arr3)

#print(arr)
#print(B)
#print(X)
df=pd.DataFrame({"A1":arr1,"A2":arr2,"A3":arr3,"B":B})
print(df)
print(X)
```

```python
#https://www.pyimagesearch.com/2019/02/04/keras-multiple-inputs-and-mixed-data/

early_stopping_monitor=EarlyStopping(patience=3)

model.fit([arr1,arr2,arr3,B] ,X,callbacks=[early_stopping_monitor], epochs=1000)

```

```python
predictionResults=model.predict([arr1,arr2,arr3,B])

print(predictionResults)
```
