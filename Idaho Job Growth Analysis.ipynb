```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('idaho_job_growth.csv')
print(df.columns)
df['employment_2018']=df['employment_2018'].astype('float')
df['employment_2028']=df['employment_2028'].astype('float')
df['change']=df['employment_2028']-df['employment_2018']

ax=df['employment_2018'].plot()
twin_ax=ax.twinx()
df['change'].plot(ax=twin_ax,color='red')
```

```python
grouped_2018=df.groupby('Industry')['employment_2018'].mean()

fig,ax=plt.subplots(figsize=(8,30))
sns.barplot(y=grouped_2018.index,x=grouped_2018.sort_values(ascending=False).values,ax=ax)
plt.show()
```
