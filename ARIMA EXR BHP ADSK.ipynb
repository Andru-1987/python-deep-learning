```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from numpy import cumsum
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df_exr_dividends=pd.read_csv('EXR_dividend.csv',parse_dates=['Date']).dropna()
df_exr_closing_price=pd.read_csv('EXR.csv',parse_dates=['Date']).dropna()

df_bhp_dividends=pd.read_csv('BHP_dividend.csv',parse_dates=['Date']).dropna()
df_bhp_closing_price=pd.read_csv('BHP.csv',parse_dates=['Date']).dropna()

df_adsk_closing_price=pd.read_csv('ADSK.csv',parse_dates=['Date']).dropna()


df_exr_dividends.set_index('Date',inplace=True)
df_exr_closing_price.set_index('Date',inplace=True)

df_bhp_dividends.set_index('Date',inplace=True)
df_bhp_closing_price.set_index('Date',inplace=True)

df_adsk_closing_price.set_index('Date',inplace=True)

fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(df_exr_dividends, marker='o')
ax1.set_ylabel('Dividends')

ax2.plot(df_exr_closing_price)
ax2.set_ylabel('Closing Price')
plt.show()

fig, (ax1, ax2) = plt.subplots(2)

ax1.plot(df_bhp_dividends, marker='o')
ax1.set_ylabel('Dividends')

ax2.plot(df_bhp_closing_price)
ax2.set_ylabel('Closing Price')
plt.show()

df_adsk['Close'].plot()
plt.title("ADSK")
plt.show()
```

```python
def plotARIMA(df,order,start,label):
    model=ARIMA(df, order=order)
    results=model.fit()
    forecast=results.get_prediction(start=start)
    mean_forecast=forecast.predicted_mean
    confidence_intervals=forecast.conf_int()
    lower_limit = confidence_intervals['lower '+label]
    upper_limit = confidence_intervals['upper '+label]
    plt.plot(df.index,df,color='green',label='NEE')
    plt.plot(mean_forecast.index, mean_forecast.values, color='red',label='forecast')
    plt.fill_between(lower_limit.index,lower_limit, upper_limit, color='pink')
    plt.show()

plotARIMA(df_exr_dividends, (2,0,1),-5,"Dividends")    
plotARIMA(df_bhp_dividends, (2,0,1),-5,"Dividends")
```

```python

    
plotARIMA(df_exr_closing_price['Close'], (2,0,0),-25,"Close")
plotARIMA(df_bhp_closing_price['Close'], (2,0,0),-25,"Close")
plotARIMA(df_adsk_closing_price['Close'], (2,0,0),-25,"Close")
```

```python
# Make arma forecast of next 10 differences
model=ARIMA(df_exr_closing_price['Close'], order=(2,0,0))

arma_results=model.fit()

arma_diff_forecast = arma_results.get_forecast(steps=10).predicted_mean

# Integrate the difference forecast
arma_int_forecast = np.cumsum(arma_diff_forecast)

# Make absolute value forecast
arma_value_forecast = arma_int_forecast + df_exr_closing_price.iloc[-1,0]

# Print forecast
#print(arma_value_forecast)

arma_value_forecast.plot()
plt.show()
```
