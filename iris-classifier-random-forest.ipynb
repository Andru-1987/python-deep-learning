```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from collections import Counter

#features: petal length, petal width, sepal length, sepal width
#target species: versicolor, virginica, setsoa
plt.style.use('ggplot')

np.random.seed(0)

iris=load_iris()

#bunch lets you use a python dict like an object
print(type(iris))
print(iris.keys())
#feature and target are numpy arrays

print('features type:' + str(type(iris.data)))
print('target type:' + str(type(iris.target)))
print(iris.data.shape)

df=pd.DataFrame(iris.data,columns=iris.feature_names)

print(df.head())

df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)

#print(df['species'])

df['is_train']=np.random.uniform(0,1,len(df))<=.75

train, test=df[df['is_train']==True], df[df['is_train']==False]

print("Number of training observations",len(train))
print("Nmber of test observations",len(test))

print(df.head())

features = df.columns[:4]

print("features")
print(features)

y=pd.factorize(train['species'])[0]
#y=iris.target

#print(y)

random_forest=RandomForestClassifier(random_state=0,n_jobs=2, n_estimators=10,verbose=1)

random_forest.fit(train[features],y)

print(random_forest)

Y_prediction=random_forest.predict(test[features])
prediction=iris.target_names[Y_prediction]
print(prediction)

print(Counter(prediction))

print(random_forest.predict_proba(test[features]))

print(pd.crosstab(test['species'],prediction,rownames=['actual species'],colnames=['predicted species']))

acc_random_forest= round(random_forest.score(train[features], y) * 100, 2)


results = pd.DataFrame({
    'Model': ['Decision Tree'],
    'Score': [acc_random_forest]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
print(result_df.head(9))
```
